```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
trdf= pd.read_csv('/content/sample_data/train.csv')
tedf= pd.read_csv('/content/sample_data/test.csv')
trdf.head()
```

```python
trdf.drop(['PassengerId', 'Name', 'Ticket', 'Cabin','Fare'], axis=1, inplace=True)


trdf['Age'].fillna(trdf['Age'].median(), inplace=True)
trdf['Embarked'].fillna(trdf['Embarked'].mode()[0], inplace=True)

label_encoder = LabelEncoder()
trdf['Sex'] = label_encoder.fit_transform(trdf['Sex'])
trdf['Embarked'] = label_encoder.fit_transform(trdf['Embarked'])


tedf['Age'].fillna(trdf['Age'].median(), inplace=True)
tedf['Embarked'].fillna(trdf['Embarked'].mode()[0], inplace=True)

label_encoder = LabelEncoder()
tedf['Sex'] = label_encoder.fit_transform(tedf['Sex'])
tedf['Embarked'] = label_encoder.fit_transform(tedf['Embarked'])
```

```python
train_y=trdf['Survived']#dependent variable
train_x=trdf[['Embarked','Parch', 'Pclass', 'Age', 'SibSp','Sex']]#independent variables
```

```python
train_y.shape
```

```python
train_x.shape
```

```python

test_x=tedf[['Embarked','Parch', 'Pclass', 'Age', 'SibSp','Sex']]#independent variables
```

```python
test_x.shape
```

```python
model = LogisticRegression()
model.fit(train_x, train_y)
```

```python
y_pred = model.predict(train_x)


# Model Evaluation
accuracy = accuracy_score(train_y, y_pred)
print("Accuracy:", accuracy)

```
