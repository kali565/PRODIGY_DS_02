```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
# Load the Titanic dataset
titanic_df = pd.read_csv('/content/titanic.csv')

```

```python
# 1. Check for missing data
missing_data = titanic_df.isnull().sum()
# Display missing data counts for each column
print("Missing Data:")
print(missing_data)

```

```python
# 2. Explore basic statistics of the dataset
summary_statistics = titanic_df.describe()
# Display summary statistics
print("\nSummary Statistics:")
print(summary_statistics)

```

```python
# 3. Visualize the data to identify patterns and trends

# a. Survived
sns.countplot(data=titanic_df, x='Survived')
plt.title('Survival Count')
plt.show()

```

```python
# b. Age
plt.figure(figsize=(8, 4))
sns.histplot(data=titanic_df, x='Age', bins=20, kde=True)
plt.title('Age Distribution')
plt.show()

```

```python
# c. Pclass
sns.countplot(data=titanic_df, x='Pclass', hue='Survived')
plt.title('Survival by Pclass')
plt.show()

```

```python
# d. Sex
sns.countplot(data=titanic_df, x='Sex', hue='Survived')
plt.title('Survival by Sex')
plt.show()

```

```python
# e. Embarked
sns.countplot(data=titanic_df, x='Embarked', hue='Survived')
plt.title('Survival by Embarked')
plt.show()

```

```python
# f. SibSp (Number of Siblings/Spouses)
sns.countplot(data=titanic_df, x='SibSp', hue='Survived')
plt.title('Survival by SibSp')
plt.show()

```

```python
# g. Parch (Number of Parents/Children)
sns.countplot(data=titanic_df, x='Parch', hue='Survived')
plt.title('Survival by Parch')
plt.show()
```
